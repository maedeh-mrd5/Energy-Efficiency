```python
import numpy as np
import matplotlib.pyplot as plt

y = 12

X = [3, 2]

W = np.zeros(2)

lr = 0.01

n = len(W)

epoc = 50

cost = np.zeros(epoc)

min_value = 0.06

number_epoc = 0

array_of_W0 = np.zeros(epoc)

array_of_W1 = np.zeros(epoc)

for e in range(epoc):
    y_model = np.dot(X, W)
    error = y_model - y
    
    last_w = [W[0], W[1]]
    
    W[0] = W[0] - lr * error * X[0] 
    
    W[1] = W[1] - lr * error * X[1] 
        
    cost[e] = error ** 2
    
    array_of_W0[e] = W[0]
    
    array_of_W1[e] = W[1]
    
    if abs(W[0] - last_w[0]) < min_value:
        break
    
    number_epoc += 1

    print(W)

print("-------------------------------")        
print(min(cost))
        
print("-------------------------------")        
print(W[0])
print(W[1])

print("-------------------------------") 
print(number_epoc)       

print("-------------------------------") 
plt.scatter(array_of_W0, cost)
plt.scatter(array_of_W1, cost)
plt.show()

    
```
